### Segmentaion Analysis

```python
low_threshold = data['Daily_Minutes_Spent'].quantile(0.25)
high_threshold = data['Daily_Minutes_Spent'].quantile(0.75)

print(f"Threshold values :-  Low {low_threshold} , High {high_threshold}")
```

```python
def label_enagement(minutes):
    if minutes <= low_threshold:
        return 'Low Engagement'
    elif minutes <= high_threshold:
        return 'Moderate Engagement'
    else:
        return 'High Engagement'
    
data['Engagement_Level'] = data['Daily_Minutes_Spent'].apply(label_enagement)
data.head()
```

```python
time_threshold_high = data['Daily_Minutes_Spent'].quantile(0.75)
time_threshold_low = data['Daily_Minutes_Spent'].quantile(0.25)

posts_threshold_high = data['Posts_Per_Day'].quantile(0.75)
posts_threshold_low = data['Posts_Per_Day'].quantile(0.25)

print(f"Time Threshold values :-  Low {time_threshold_low} , High {time_threshold_high}")
print(f"Posts Threshold values :-  Low {posts_threshold_low} , High {posts_threshold_high}")
```

```python
def label_engagement_pattern(minutes, posts):
    if minutes >= time_threshold_high and posts <= posts_threshold_low:
        return 'High Time, Low Posts'
    elif minutes >= time_threshold_high and posts >= posts_threshold_high:
        return 'High Time, High Posts'
    elif minutes <= time_threshold_low and posts >= posts_threshold_high:
        return 'Low Time, High Posts'
    elif minutes <= time_threshold_low and posts >= posts_threshold_low:
        return 'Low Time, Low Posts'
    else:
        return 'Moderate Engagement'
    
data['Engagement_Pattern'] = data.apply(lambda row: label_engagement_pattern(row['Daily_Minutes_Spent'], row['Posts_Per_Day']), axis=1)
data.head(15)
```

```python
correlation = data['Daily_Minutes_Spent'].corr(data['Posts_Per_Day'])

print(f"Correlation {correlation}")
```

### Advanced Analytics

```python
features = data[['Daily_Minutes_Spent', 'Posts_Per_Day', 'Likes_Per_Day', 'Follows_Per_Day']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
inertia = []
for k in range(1, 11):
    Kmean = KMeans(n_clusters=k, n_init=10, random_state=44)
    Kmean.fit(scaled_features)
    inertia.append(Kmean.inertia_)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow method for Optical K")
plt.xlabel("Number Of Clusters")
plt.ylabel('inertia')
```

```python
k = 5
kmean = KMeans(5, n_init=10, random_state=44)
data['Cluster'] = kmean.fit_predict(scaled_features)

plt.figure(figsize=(10,6))
sns.scatterplot(x='Daily_Minutes_Spent', y='Likes_Per_Day', hue='Cluster', data=data, palette='rocket')
plt.title("Users Clusters Based On Daily Minutes Spent And Likes Per Day")
plt.xlabel("Daily Minutes Spent")
plt.ylabel("Likes Per Day")
plt.show()
```
